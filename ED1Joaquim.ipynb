Importando pacotes e dando upload no arquivo lena

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


from zipfile import ZipFile
from urllib.request import urlretrieve

from IPython.display import Image
```

Nomeando a imagem e testando

```python
arquivo = 'lena.pgm'

lena = cv2.imread(arquivo, 0) 
   
plt.imshow(lena, cmap='gray')
plt.show()
```

Selecionando a rea do olho para saber onde trocar depois, e a rea acima do olho que servir como a piscada

```python
olho = lena[257:276, 250:280]
olhoc = lena[250:260, 255:280]

plt.imshow(olho,  cmap='gray')
plt.show()
plt.imshow(olhoc,  cmap='gray')
plt.show()
```

Aumentando o tamanho da imagem acima do olho para tapar o olho original

```python
xolho, yolho = 38, 23
olhoc2 = cv2.resize(olhoc, (xolho, yolho))
plt.imshow(olhoc2, cmap='gray')
plt.show()
```

Definindo a imagem com o olho fechado

```python
lena1 = lena.copy()
lena1[252:275, 250:288] = olhoc2
plt.imshow(lena1, cmap='gray')
plt.show()
```

Criando a animao das 3 piscadas

```python
def piscar(n):
    for i in range(n):
        cv2.imshow('Piscando',lena)
        cv2.waitKey(300)
        
        cv2.imshow('Piscando', lena1)
        cv2.waitKey(170)

        cv2.imshow('Piscando',lena)
piscar(3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
