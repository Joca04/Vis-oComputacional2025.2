# Importando pacotes e abrindo a imagem

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
arquivo = 'lena.pgm'

lena = cv2.imread(arquivo, 0) 
lenad = lena.astype(np.float64)

plt.imshow(lena, cmap='gray')
plt.show()
```

# Aumentando o Brilho

```python
brilho = 100
lenab = cv2.convertScaleAbs(lena, beta = brilho)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lena, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Brilho")
plt.imshow(lenab, cmap='gray')

plt.show()
```

# Aumentando o Contraste

```python
contraste = 2.1
lenac = cv2.convertScaleAbs(lena, alpha= contraste)

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lena, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Contraste")
plt.imshow(lenac, cmap='gray')

plt.show()
```

# Tornando a imagem Negativa

```python
lenan = 255 - lena

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(lena, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Negativo")
plt.imshow(lenan, cmap='gray')

plt.show()
```

# Posterizando a imagem

```python
n = 12
fator = 256 / n
poster = (lena // fator) * fator
poster = poster.astype(np.uint8)

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(lena, cmap='gray')

plt.subplot(1,2,2)
plt.title(f"Poster ({n} nveis)")
plt.imshow(poster, cmap='gray')

plt.show()
```
