# Importando pacote e opencv

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Lendo a imagem 'penguins'

```python
arquivo = 'penguins.png'
peng = cv2.imread(arquivo)
plt.imshow(peng, cmap='gray')
plt.show()
```

# Plotando o histograma da imagem original

```python
hist = cv2.calcHist([peng], [0], None, [256], [0, 256])
plt.title("Histograma")
plt.plot(hist, color='black')
```

# Calculando a cor que mais aparece

```python
maxim = np.argmax(hist)
print(maxim)
```

# Definindo limite de transio para binrio:
### Analisando a imagem, 100 seria um limite interessante para tornar essa imagem em binria, j que  logo aps a grande cada na frequncia dos cinzas mais escuros

# Plotando imagem binria e histograma dela

```python
ret, binaria = cv2.threshold(peng, 100, 255, cv2.THRESH_BINARY)
histb = cv2.calcHist([binaria], [0], None, [256], [0, 256])
plt.imshow(binaria,cmap='grey')
plt.show()
plt.title("Histograma binrio")
plt.plot(histb, color='gray')
```
