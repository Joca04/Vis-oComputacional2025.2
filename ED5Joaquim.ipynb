# Importando pacotes

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Aplicando algoritmo de subtração para diferença de frames 

```python
video = cv2.VideoCapture('traffic_sequence.mpg')
ret, frame_anterior = video.read()
frame_anterior_gray = cv2.cvtColor(frame_anterior, cv2.COLOR_BGR2GRAY)
while True:
    ret, frame = video.read()
    if not ret:
        break
        
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    diferenca = cv2.absdiff(frame_anterior_gray, gray)
    
    cv2.imshow('Diferenca (Movimento)', diferenca)
    cv2.imshow('Original', frame)
    
    frame_anterior_gray = gray.copy()

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
        
video.release()
cv2.destroyAllWindows()
```
