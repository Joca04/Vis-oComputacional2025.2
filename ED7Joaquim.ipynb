# Importando pacotes

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#  Lendo o arquivo e convertendo para RGB

```python
arquivo = 'flowers9.png'
flowers = cv2.imread(arquivo)
florgb = cv2.cvtColor(flowers, cv2.COLOR_BGR2RGB)
```

# Montando os histogramas

```python
plt.figure(figsize=(20, 14))

histR = cv2.calcHist([florgb], [0], None, [256], [0,256])
plt.subplot(2, 2, 1)
plt.title('Histograma Vermelho')
plt.plot(histR, color = 'red')

histG = cv2.calcHist([florgb], [1], None, [256], [0,256])
plt.subplot(2, 2, 2)
plt.title('Histograma Verde')
plt.plot(histG, color = 'green')

histB = cv2.calcHist([florgb], [2], None, [256], [0,256])
plt.subplot(2, 2, 3)
plt.title('Histograma Azul')
plt.plot(histB, color = 'blue')

plt.show()
```

# Separando a imagem nos 3 espectros e plotando cada um

```python
r, g, b = cv2.split(florgb)

plt.figure(figsize=(20, 14))

plt.subplot(2, 2, 1)
plt.title('RGB')
plt.imshow(florgb)

plt.subplot(2, 2, 2)
plt.title('Vermelho')
plt.imshow(r)

plt.subplot(2, 2, 3)
plt.title('Verde')
plt.imshow(g)

plt.subplot(2, 2, 4)
plt.title('Azul')
plt.imshow(b)
plt.show()
```
